```python
#importing relevant libraries
import numpy as np #random number generation
import matplotlib.pyplot as plt #to plot graphs

#Parameters
S0 = 100 #initial stock price --> £100
mu = 0.05 #expected return --> 5% average annual return 
sigma = 0.2 #volatility --> 20%
T = 1 #time horizon --> 1 year
dt = 1/252 #daily steps, 252 trading days/year
N = int(T/dt) #number of steps
M = 10000 #number of simulations

paths = np.zeros((N+1, M)) #matrix created
paths[0] = S0 #start from S0 variable

for m in range (M):
    for t in range(1, N+1):
        z = np.random.normal()
        paths[t, m] = paths[t-1, m] * np.exp((mu - 0.5 * sigma**2)*dt + sigma*np.sqrt(dt)*z)

plt.figure(figsize=(10,6))
plt.plot(paths)
plt.title('Monte Carlo Stock Price Simulation')
plt.xlabel('Days')
plt.ylabel('Price')
plt.show()
```
